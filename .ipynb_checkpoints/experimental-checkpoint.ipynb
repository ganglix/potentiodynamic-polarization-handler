```python
%matplotlib notebook
from Tafel_LPR_fit_notebook_IRcomp_IFilm import*
```

```python
# read excel and rename columns. Can use differnt data source as long as you get df 'I' 'E'
# I is signed. negative for cathodic current
df = pd.read_excel('data.xlsx') # first column I, second E


# Create fit object
fit = Tafit(
                Info(
                    'use data frame', # use filename as description
                    pd_dfIE = df,     # assign df
                    use_pd_df = True, # set to true to use df to bypass the hassle detecting cathodic current from absolute currents
                    area=1.0          # current density = current / 1 
                    )
            )   


fit.BV_LPR_interact(anodic_range=0.15,cathodic_range=0.15)
```

```python
# see all attributes of fit
vars(fit)
```

```python
# display fitted results
print(fit.result)

print()
# individual one for full percision
print(fit.result.Icorr)


# Icorr_LPR should be not too far from Icorr. Here Icorr_LPR was calculated use small range near OCP as if it is a LPR test.
# Strictlly, LPR is not valid here, but we use it to cross check.
```
